# Dicionários

```python
# Criando dicionário do Exemplo
idades = {
    "joao": 10,
    "roberto": 16,
    "ana": 13,
    "nicolas": 9,
    "miguel": 12
}
```

```python
# Qual a idade do João?
print(idades["joao"])
```

```python
# Loop num dicionário (só com chave)
for nome in idades:
  print(nome, idades[nome])
```

```python
# Loop num dicionário (com chaves e valores)
for nome, idade in idades.items():
  print(nome, idade)
```

```python
# Criando tabela com dicionário
tabela = {
    "Nome": ["Roberto", "Ana", "Miguel", "Nicolas", "Eduardo"],
    "Idade": [19, 22, 24, 18, 21],
    "Curso": ["Informática", "Biologia", "Engenharia", "Direito", "Medicina"]
}
```

```python
print(tabela["Nome"][1])
```

```python
# acessando terceiro linha
linha = 2
for coluna in tabela:
  print(tabela[coluna][linha])
```

```python
# media das idades
import statistics as stat
media = stat.mean(tabela["Idade"])
print(media)
```

# Pandas

```python
# importando a biblioteca
import pandas as pd
```

```python
# Criando um dataframe a partir de uma tabela
tabela = {
    "Nome": ["Roberto", "Ana", "Miguel", "Nicolas", "Eduardo"],
    "Idade": [19, 22, 24, 18, 21],
    "Curso": ["Informática", "Biologia", "Engenharia", "Direito", "Medicina"]
}
df = pd.DataFrame(data=tabela)
```

```python
# Visualizando a tabela
df
```

```python
# Acessando uma coluna específica
df["Nome"]
```

```python
# Acessando algumas colunas
df[["Nome", "Curso"]]
```

```python
# Calculando estatísticas
media = df["Idade"].mean()
print(media)
```

```python
# Filtragem
df[df["Idade"] > 20]
```

```python
# numero de linhas
print(len(df))
```

```python
# numero de colunas
print(len(df.columns))
```

# Exercícios

```python
# criando um data frame para usar nos exercícios
# esse data frame contem dados de pacientes que tiveram covid
dados = {
    "Codigo": [40, 354, 123, 904, 231, 999, 506, 908, 100, 432],
    "Idade": [19, 15, 13, 21, 40, 50, 11, 61, 38, 27],
    "Sexo": ["M", "F", "M", "M", "M", "F", "M", "M", "F", "F"],
    "Tempo Internado": [17, 22, 24, 21, 23, 5, 19, 12, 24, 27],
    "Óbito": ["S", "N", "N", "N", "S", "S", "N", "N", "N", "N"]
}
df = pd.DataFrame(data=dados)
df
```

```python
# 1) Quantos pacientes morreram?
numero_mortos = len(df[df["Óbito"] == "S"])
print(numero_mortos)
```

```python
# 2) Qual o tempo médio de internação?
media = df["Tempo Internado"].mean()
print(media)
```

```python
# 3) Qual a idade média dos homens?
homens = df[df["Sexo"] == "M"]
media = homens["Idade"].mean()
print(media)
```

```python
# 4) Quando tempo, em média, os que sobreviveram ficaram internados?
sobreviventes = df[df["Óbito"] == "N"]
media = sobreviventes["Tempo Internado"].mean()
print(media)
```

```python
# 5) Mostre a tabela com apenas as mulheres com mais de 30 anos
mulheres = df[df["Sexo"] == "F"]
mulheres_mais30 = mulheres[mulheres["Idade"] > 30]
mulheres_mais30
```
