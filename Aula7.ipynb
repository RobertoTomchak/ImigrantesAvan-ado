# Pandas

```python
# importando a biblioteca
import pandas as pd
```

```python
# Criando um dataframe a partir de uma tabela
tabela = {
    "Nome": ["Roberto", "Ana", "Miguel", "Nicolas", "Eduardo"],
    "Idade": [19, 22, 24, 18, 21],
    "Curso": ["Informática", "Biologia", "Engenharia", "Direito", "Medicina"]
}
df = pd.DataFrame(data=tabela)
```

```python
# Visualizando a tabela
df
```

```python
# Acessando uma coluna específica
df["Nome"]
```

```python
# Acessando algumas colunas
df[["Nome", "Curso"]]
```

```python
# Calculando estatísticas
media = df["Idade"].mean()
print(media)
```

```python
# Filtragem
df[df["Idade"] > 20]
```

```python
# Podemos ver o tamanho de uma tabela também
len(df)
```
