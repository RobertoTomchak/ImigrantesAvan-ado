# Analisando dados

```python
# 1) Qual a idade média dos candidatos?
print(df["NU_IDADE"].mean())
```

```python
# 2) Quantos candidatos são de Curitiba?
#    Salve em um arquivo apenas os dados dos curitibanos
df_curitiba = df[df["NO_MUNICIPIO_ESC"] == "Curitiba"]
tam = len(df_curitiba)
print(tam)
df_curitiba.to_csv("dados_curitiba.csv")
```

```python
# 3) Quantos estrangeiros fazem o ENEM?
df_estrangeiros = df[df["TP_NACIONALIDADE"] == "Estrangeiro(a)"]
print(len(df_estrangeiros))
```

```python
# 4) Qual a porcentagem de alunos de escola pública?
df_publica = df[df["TP_ESCOLA"] == "Publica"]
num_publica = len(df_publica)
num_total = len(df)
porcentagem = num_publica / num_total * 100
print(f"{round(porcentagem, 2)}%")
```

```python
# 5) Calcule a nota média de cada candidato
df["NOTA_MEDIA"] = (df["NU_NOTA_CN"] + df["NU_NOTA_LC"] + df["NU_NOTA_CH"] + df["NU_NOTA_MT"] + df["NU_NOTA_REDACAO"]) / 5
df
```

```python
# 6) Qual a nota média dos estudantes de escola pública?
df_publica = df[df["TP_ESCOLA"] == "Publica"]
media = df_publica["NOTA_MEDIA"].mean()
print(media)
```

```python
# 7) Qual a nota média na redação na Região Sul?
df_sul = df[df["SG_UF_ESC"].isin(["PR", "SC", "RS"])]
media_redacao_sul = df_sul["NU_NOTA_REDACAO"].mean()
print(media_redacao_sul)
```

```python
# 8) A nota média da redação no Sul é menor que no
# Brasil como um todo?
media_redacao_brasil = df["NU_NOTA_REDACAO"].mean()
print(media_redacao_brasil)

if media_redacao_sul > media_redacao_brasil:
  print("Sul foi melhor na redacao")
else:
  print("Brasil é melhor que o Sul na redacao")
```

```python
# 9) Qual a proporção das mulheres no ENEM?
#    Elas tendem a ter notas melhores que os homens?
df_mulheres = df[df["TP_SEXO"] == "F"]
df_homens = df[df["TP_SEXO"] == "M"]
proporcao_mulheres = len(df_mulheres) / len(df) * 100
print(proporcao_mulheres)

nota_mulheres = df_mulheres["NOTA_MEDIA"].mean()
nota_homens = df_homens["NOTA_MEDIA"].mean()
print(nota_mulheres, nota_homens)
if nota_mulheres > nota_homens:
  print("Mulheres tiraram maiores notas")
else:
  print("Homens tiraram notas maiores")
```

```python
# 10) A nota de corte de Medicina é 807,20.
# Quantos alunos passariam em Medicina?
nota_corte = 807.20
df_passariam = df[df["NOTA_MEDIA"] >= nota_corte]
print(len(df_passariam))
```
